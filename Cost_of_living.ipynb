```python
import pandas as pd

# قراءة البيانات
df = pd.read_excel("/content/cost_of_living.xlsx")

# حساب تكلفة المعيشة
df["total_cost"] = df["rent"] + df["food"] + df["transport"]

# ترتيب المدن من الأرخص للأغلى
df_sorted = df.sort_values("total_cost")

print("🏙️ Cheapest to Expensive Cities:")
print(df_sorted[["city", "total_cost"]])

# مقارنة الراتب مع التكلفة
df["ratio"] = df["salary_index"] / df["total_cost"]

print("\n💰 Best Cities (Salary vs Cost):")
print(df[["city", "ratio"]].sort_values("ratio", ascending=False))
```

```python
import matplotlib.pyplot as plt

plt.bar(df["city"], df["total_cost"])
plt.title("Cost of Living by City")
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_excel("cost_of_living.xlsx")

df["total_cost"] = df["rent"] + df["food"] + df["transport"]

plt.figure(figsize=(10,5))
plt.bar(df["city"], df["total_cost"])
plt.title("Cost of Living by City")
plt.xlabel("City")
plt.ylabel("Total Cost")
plt.xticks(rotation=45)
plt.show()
```

```python
df["ratio"] = df["salary_index"] / df["total_cost"]

print("💰 Best cities based on salary vs cost:")
print(df[["city", "ratio"]].sort_values("ratio", ascending=False))
```
